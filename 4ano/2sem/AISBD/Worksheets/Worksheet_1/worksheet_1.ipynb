## Questão 5 - Indique, quando possível, o significado da variável.

- **tpep_pickup_datetime**: Data e hora de início da viagem.

- **tpep_dropoff_datetime**: Data e hora de término da viagem.

- **passenger_count**: Número de passageiros na viagem.

- **trip_distance**: Distância percorrida durante a viagem, medida em milhas pelo taxímetro.

- **RatecodeID**: The final rate code in effect at the end of the trip.
    -   1 = Standard rate
    -   2 = JFK
    -   3 = Newark
    -   4 = Nassau or Westchester
    -   5 = Negotiated fare
    -   6 = Group ride
    -   99 = Null/unknown
- **store_and_fwd_flag**: Indica se os dados da viagem foram armazenados e encaminhados posteriormente (Y) ou se foram transmitidos em tempo real (N).

- **PULocationID**: Código do local de partida da viagem.

- **DOLocationID**: Código do local de destino da viagem.

- **payment_type**: Método de pagamento utilizado para a viagem (0 : Flex Fare trip, 1: cartão de crédito, 2: dinheiro, 3: sem pagamento, 4: dispute, 5: unknown, 6: voided trip).

- **fare_amount**: Valor da tarifa cobrada pela viagem, em dólares.

- **extra**: Taxas adicionais cobradas durante a viagem.

- **mta_tax**: Taxa fixa cobrada pela MTA (Metropolitan Transportation Authority) para viagens que ocorrem dentro da cidade de Nova York.

- **tip_amount**: Valor da gorjeta dada ao motorista, em dólares.

- **tolls_amount**: Valor total das taxas de pedágio cobradas durante a viagem, em dólares.

- **improvement_surcharge**: Taxa adicional cobrada para melhorias na infraestrutura de transporte, em dólares.

- **total_amount**: Valor total pago pelo passageiro, incluindo tarifa, taxas adicionais e pedágios, em dólares.

- **congestion_surcharge**: Taxa adicional cobrada durante períodos de congestionamento, em dólares.

- **airport_fee**: Taxa adicional cobrada para viagens que ocorrem dentro do aeroporto, em dólares.

- **cbd_congestion_fee**: Per-trip charge for MTA's Congestion Relief Zone starting Jan. 5, 2025

## Questão 14 - Calcule a duração da viagem (em minutos), utilizando as variáveis temporais disponíveis, e analise a distribuição dessa duração.

```python
df['trip_time_in_minutes'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60
# print(f"Valores mínimos e máximos respetivamente para 'trip_time_in_minutes': \n Minimo {df['trip_time_in_minutes'].min()}\n Maximo {df['trip_time_in_minutes'].max()}\n")
# print(f"Percentis respetivamente para 'trip_time_in_minutes':\n{df['trip_time_in_minutes'].quantile([0.01, 0.5, 0.99])}")
```

## Questão 15 - Identifique viagens cuja duração seja:
a. negativa

b. nula

c. excessivamente elevada.

```python
## a. negativa
negative_duration = df[df['trip_time_in_minutes'] < 0]
print(f"Viagens com duração negativa: {len(negative_duration)}")

## b. nula
null_duration = df[(df['trip_time_in_minutes'] == 0) | (df['trip_time_in_minutes'].isnull()) | (df['trip_time_in_minutes'] == '')]
print(f"Viagens com duração nula: {len(null_duration)}")

## c. excessivamente elevada.
excessive_duration = df[df['trip_time_in_minutes'] > 300]  # Considerando viagens com duração superior a 5 horas como excessivamente elevadas
print(f"Viagens com duração excessivamente elevada: {len(excessive_duration)}")
```
